````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
base = pd.read_csv('Profissao- Cientista de Dados_M10_support material.csv')
base.head()
````

````python
base2 = base.drop(['Unnamed: 0', 'index', 'mau'], axis=1)
base2.head()
````

2) Avalie um gráfico de dispersão (*scatterplot*) específico para as duas variáveis mais correlacionadas com *renda*.

````python
_ = sns.regplot(x='tempo_emprego', y='renda', data = base2)
base2.corr()
````

````python
_ = sns.regplot(x='posse_de_veiculo', y='renda', data = base2)
base2.corr()
````

3) Na sua opinião, há outliers na variável renda?

Sim, ambas as correlações possuem outliers.

4) Calcule o logaritmo da variável renda e repita a sequência de análise

````python
base2_log = base2.copy() 
````

````python
base2_log['log_renda'] = np.log(base2_log['renda'])

_ = sns.regplot(x='tempo_emprego', y='log_renda', data = base2_log)
````

````python
base2_log['log_renda'] = np.log(base2_log['renda'])

_ = sns.regplot(x='posse_de_veiculo', y='log_renda', data = base2_log)
````

5) A aplicação do LOG você acha que melhorou ou piorou a análise?

Em ambos os caso, a aplicação do LOG melhorou muito a avaliação gráfica das correlações das variáveis.
