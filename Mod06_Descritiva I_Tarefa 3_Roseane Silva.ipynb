```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
sinasc = pd.read_csv('SINASC_RO_2019.csv')
sinasc.head()
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
sinasc.loc[sinasc['APGAR5']<=3, 'APGAR5_cat'] =  'asfixia grave'
sinasc.loc[(sinasc['APGAR5'] >= 4) & (sinasc['APGAR5'] <= 5), 'APGAR5_cat'] =  'asfixia moderada'
sinasc.loc[(sinasc['APGAR5'] >= 6) & (sinasc['APGAR5'] <= 7), 'APGAR5_cat'] =  'asfixia leve'
sinasc.loc[(sinasc['APGAR5'] >= 8), 'APGAR5_cat'] =  'normal'
```

```python
sinasc.columns
```

```python
sinasc['APGAR5_cat'].value_counts()
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
sns.boxplot(x="APGAR5", y="PESO", hue = 'SEXO', data=sinasc)
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
ges_ap5 = pd.crosstab(sinasc['GESTACAO'], sinasc['APGAR5'])
ges_ap5
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
ges_ap5.sum(axis = 1)
```

```python
ges_ap5_pctlin = ges_ap5.div(ges_ap5.sum(axis = 1), axis = 0)
ges_ap5_pctlin
```

```python
ges_ap5_pctlin.sum(axis = 1)
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
ges_ap5_pctlin.plot.bar()
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
```

#### 7) Veja se há relação entre APGAR5 e o peso do bebê. Isso pode ser feito de diversas formas.
```python
ax = sns.boxplot(x="APGAR5", y="PESO", data=sinasc)
plt.title("APGAR5 vs. PESO")
plt.xlabel("APGAR5")
plt.ylabel("PESO (em gramas)")
```

```python
plt.scatter(sinasc['APGAR5'], sinasc['PESO'], color='b', edgecolors='black')
plt.title("APGAR5 vs. PESO")
plt.xlabel("APGAR5")
plt.ylabel("PESO (em gramas)")
plt.grid(True)

plt.show()
```

A partir de uma análise gráfica dos dois exemplos acima, pode-se concluir que tais váriavés não possuem uma relação direta. Porém alguns pontos importantes podem ser abordados:
1. A maior parte da amostra apresentou o APGAR5 desejado (maior ou igual a 4);
2. Para a amostra que teve APGAR5 menor que o esperado (4), com destaque para os valores 0 e 2, as amostras possuem mediana das pesagens inferiores a 1000g, que é uma condição de pesagem crítica (provavelmente esses bêbes tiveram uma má formação ou nasceram muito antes do esperado, até mesmo para bêbes prematuros);
3. Para a amostra que teve um APGAR5 entre 5 e 10, a mediana dessas amostras possuem uma normalidade, apresentando valores de pesagem em sua maioria superior a 3000g.
