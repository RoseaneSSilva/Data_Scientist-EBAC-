### 3. Qual o municipio que nasceu menos bebe em 2019?
    - qual a idade media, maxima, minima das maes nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?

```python
sinasc_raw['munResNome'].value_counts()
```

Castanheiras é a cidade com menos nascimentos em 2019, totalizando apenas 32 registros no total.

```python
info = ['mean','min','max']
condicao = sinasc_raw['munResNome'] == 'Castanheiras'

sinasc_raw[condicao].groupby('munResNome').agg({'IDADEMAE':info , 'IDADEPAI':info})
```

### 4. Qual o municipio que nasceu mais bebe no mês de março?
    - qual a quantidade de filhos vivos media, maxima, minima nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?



```python
sinasc_raw['DTNASC'] = sinasc_raw['DTNASC'].astype('datetime64')
```

```python
sinasc_raw['mes_ano'] = sinasc_raw['DTNASC'].map(lambda x: 100*x.year + x.month)
```

```python
sinasc_raw.head()
```

```python
sinasc_raw[sinasc_raw['mes_ano'] == 201903].groupby('munResNome')['IDADEMAE'].count().sort_values(ascending=False)
```

No mês de março de 2019, Porto Velho totalizou 744 registros de nascimentos, sendo a cidade que mais concentrou nascimentos nesse período.

```python
condicao2 = (sinasc_raw['mes_ano'] == 201903) & (sinasc_raw['munResNome'] == 'Porto Velho')
```

```python
sinasc_raw[condicao2].groupby('munResNome').agg({'QTDFILVIVO': info, 'IDADEMAE':info , 'IDADEPAI':info})
```

Para esse tópico, foram analisados critérios relacionados a cor/raça das mães que deram a luz no ano de 2019 em Rondonia.        
Raça/Cor:
1- Branca
2- Preta
3- Amarela
4- Parda
5- Indígena

```python
sinasc_raw['RACACORMAE'].value_counts()
```

```python
sinasc_raw.groupby(['DTNASC', 'RACACORMAE'])['IDADEMAE'].count().unstack().plot(figsize=[15,10]);
plt.ylabel('quantidade de nascimentos por raças');
```

Analisando o gráfico acima, pode-se afirmar que a maioria das mães que tiveram filhos em Rondonia em 2019, se declararam como pardas e brancas, enquanto que as que se declararam pretas e índigenas foram minoria.
