````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
base = pd.read_csv('Profissao- Cientista de Dados_M10_support material.csv')
base.head()
````

````python
base2 = base.drop(['Unnamed: 0', 'index', 'mau'], axis=1)
base2.head()
````

1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
plt.close('all')
plt.rc('figure', figsize=(15, 10))
plt.xticks(fontsize=12,rotation=90)

ax = sns.countplot(x='data_ref', hue = 'sexo', data=base2)
ticks = ax.set_xticks(list(range(base2['data_ref'].nunique())))
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

````python
plt.close('all')
plt.rc('figure', figsize=(15, 10))
plt.xticks(fontsize=12,rotation=90)

ax = sns.countplot(x='data_ref', hue = 'posse_de_veiculo', data=base2)
ticks = ax.set_xticks(list(range(base2['data_ref'].nunique())))
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

````python
plt.close('all')
plt.rc('figure', figsize=(15, 10))
plt.xticks(fontsize=12,rotation=90)

ax = sns.countplot(x='data_ref', hue = 'posse_de_imovel', data=base2)
ticks = ax.set_xticks(list(range(base2['data_ref'].nunique())))
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

Comparando os resultados dos gráficos de perfis de média (atividade anterior) com os gráficos de barra (atividade atual), que só relacionam variáveis no decorrer do tempo, a conclusão que se tem é que a maior parte da base é constituída de dados femininos, que não possuem veículo mas possuem imóvel, e essas relações não se alteraram ao longo do tempo (15 meses). O que se poderia verificar de interessante são as relações de pessoas de  um mesmo sexo que possuem veiculo e imóvel, e sua relação com renda. 

2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
plt.close('all')
plt.rc('figure', figsize=(10, 5))
plt.xticks(fontsize=8,rotation=90)

ax = sns.pointplot(x='data_ref', y='posse_de_imovel', hue = 'sexo',
                   data=base2, dodge=True, ci = 95)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.50),ncol = 3)
````

````python
plt.close('all')
plt.rc('figure', figsize=(10, 5))
plt.xticks(fontsize=8,rotation=90)

ax = sns.pointplot(x='data_ref', y='posse_de_veiculo', hue = 'sexo',
                   data=base2, dodge=True, ci = 95)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.50),ncol = 3)
````

Nos casos em analise, pode-se afirmar que em posse de imóveis, a variável homem não é muito estavél, apresentando variações mais significativas em sua médias no decorrer do tempo. No entanto a variável posse de veiculo é muito mais estável, logo mais confiável para uma modelagem. Ambas as variáveis podem ser utilizadas numa modelagem entretanto, seria interessante relacionar posse de veículos.
